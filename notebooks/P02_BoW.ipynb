```
# Bag of Words
```

```
from src.analisis_nlp.BoW.BagOfWords import CorpusLoader, BagOfWordsAnalyzer, TFIDFAnalyzer
from src.db_manager import get_collection
```

```
col = get_collection()
```

# Carga de canciones

```
loader         = CorpusLoader(generos=["rock", "pop", "hip hop"], sample_size=20)
corpus, labels = loader.load(col)
```

# Vectores Bow

```
bow = BagOfWordsAnalyzer(max_features=50)
bow.fit_transform(corpus)
bow.summary(n_docs=3, n_words=15)
bow.demo_orthogonality(["love", "fire", "money"])
```

# TF-IDF: ponderación inteligente

```
tfidf = TFIDFAnalyzer(max_features=50)
tfidf.fit_transform(corpus)
tfidf.top_words_by_genre(labels, top_n=5)
```
